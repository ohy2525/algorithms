```python
from copy import deepcopy

n = int(input())
arr = []
for _ in range(n):
    arr.append(list(map(int, input().split())))
    
def sol(arr, x):
    #right
    if x == 0:
        for i in range(n):
            for j in range(n - 2, -1, -1):
                if arr[i][j + 1] == arr[i][j]:
                    arr[i][j + 1] += arr[i][j]
                    arr[i][j] = 0
                elif arr[i][j + 1] == 0:
                    arr[i][j + 1] = arr[i][j]
                    arr[i][j] = 0
    #left           
    if x == 1:
        for i in range(n):
            for j in range(1, n):
                if arr[i][j - 1] == arr[i][j]:
                    arr[i][j - 1] += arr[i][j]
                    arr[i][j] = 0
                elif arr[i][j - 1] == 0:
                    arr[i][j - 1] = arr[i][j]
                    arr[i][j] = 0
    #up
    if x == 2:
        for i in range(n):
            for j in range(1, n):
                if arr[j - 1][i] == arr[j][i]:
                    arr[j - 1][i] += arr[j][i]
                    arr[j][i] = 0
                elif arr[j - 1][i] == 0:
                    arr[j - 1][i] = arr[j][i]
                    arr[j][i] = 0
    #down
    if x == 3:
        for i in range(n):
            for j in range(n - 2, -1, -1):
                if arr[j + 1][i] == arr[j][i]:
                    arr[j + 1][i] += arr[j][i]
                    arr[j][i] = 0
                elif arr[j + 1][i] == 0:
                    arr[j + 1][i] = arr[j][i]
                    arr[j][i] = 0
                    
    return arr


ans = 0
def dfs(arr, cnt):
    global ans
    if cnt == 5:
        max_val = max(map(max, arr))
        ans = max(ans, max_val)
        return
    
    for i in range(4):
        copy_arr = sol(deepcopy(arr), i)
        dfs(copy_arr, cnt + 1)
        
dfs(arr, 0)

print(ans)
```
