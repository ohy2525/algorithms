https://www.acmicpc.net/problem/17413

```python
s = input()

ans = ''
tmp = ''
for i in s:
    if i == '<':
        ans += tmp[::-1]
        tmp = '<'
    elif i == '>':
        tmp += '>'
        ans += tmp
        tmp = ''
    elif i ==' ' and tmp[0] != '<':
        ans += tmp[::-1]
        ans += ' '
        tmp = ''
    else:
        tmp += i
        
ans += tmp[::-1]

print(ans)
```
