https://www.acmicpc.net/problem/17144

```python
r, c, t = map(int, input().split())
graph = []
for _ in range(r):
    graph.append(list(map(int, input().split())))
    
dx = [-1, 0, 1, 0]
dy = [0, 1, 0, -1]

def diffuse():
    check = [[0] * c for _ in range(r)]
    for x in range(r):
        for y in range(c):
            if graph[x][y]:
                tmp = 0
                for i in range(4):
                    nx = x + dx[i]
                    ny = y + dy[i]
                    if 0 <= nx  < r and 0 <= ny < c:
                        if graph[nx][ny] != -1:
                            check[nx][ny] += graph[x][y] // 5
                            tmp += graph[x][y] // 5
                
                graph[x][y] -= tmp
    for x in range(r):
        for y in range(c):
            graph[x][y] += check[x][y]

def rotate_top():
    d = 1
    before = 0
    x, y = robot_top, 1
    while True:
        nx = x + dx[d]
        ny = y + dy[d]
        if nx == r or ny == c or nx == -1 or ny == -1:
            d = (d - 1) % 4
            continue
        if x == robot_top and y == 0:
            break
        graph[x][y], before = before, graph[x][y]
        x, y = nx, ny
    
def rotate_bottom():
    d = 1
    before = 0
    x, y = robot_bottom, 1
    while True:
        nx = x + dx[d]
        ny = y + dy[d]
        if nx == r or ny == c or nx == -1 or ny == -1:
            d = (d + 1) % 4
            continue
        if x == robot_bottom and y == 0:
            break
        graph[x][y], before = before, graph[x][y]
        x, y = nx, ny

for i in range(r): 
    if graph[i][0] == -1:
        robot_top = i
        robot_bottom = i+1
        break

for _ in range(t):
    diffuse() 
    rotate_top()
    rotate_bottom()
    
answer = 2 #공기청정기  -2
for i in range(r):
    answer += sum(graph[i])
print(answer)
```
