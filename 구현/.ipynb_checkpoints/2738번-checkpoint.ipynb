https://www.acmicpc.net/problem/2738

```python
n, m = map(int, input().split())
a = []
b = []
ans = [[0 for _ in range(m)] for _ in range(n)]

for _ in range(n):
    a.append(list(map(int, input().split())))
for _ in range(n):
    b.append(list(map(int, input().split())))
    
for i in range(n):
    for j in range(m):
        ans[i][j] = a[i][j] + b[i][j]
        
for i in range(n):
    for j in range(m):
        print(ans[i][j], end = ' ')
    print()
```
