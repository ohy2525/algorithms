```python
from collections import deque

gear = []
gear.append(deque(list(map(int, input()))))
gear.append(deque(list(map(int, input()))))
gear.append(deque(list(map(int, input()))))
gear.append(deque(list(map(int, input()))))

#  
def clock(g):
    gear[g - 1].rotate(1)

def counter_clock(g):
    gear[g - 1].rotate(-1)
    
k = int(input())
for _ in range(k):
    g, d = map(int, input().split())
    
    #     
    check_list = [[] for _ in range(3)]
    for i in range(3):
        check_list[i].append(gear[i][2])
        check_list[i].append(gear[i + 1][6])
    
    #   
    if d == 1:
        clock(g)
    else:
        counter_clock(g)
        
    # 
    check_d = d * -1  #     
    if g != 1:  
        for i in range(g - 1, 0, -1):
            if check_list[i - 1][0] != check_list[i - 1][1]:
                if check_d == 1:
                    clock(i)
                    check_d = -1
                else:
                    counter_clock(i)
                    check_d = 1
            else:
                break
                
    #             
    check_d = d * -1           
    if g != 4:  
        for i in range(g - 1, 3):
            if check_list[i][0] != check_list[i][1]:
                if check_d == 1:
                    clock(i + 2)
                    check_d = -1
                else:
                    counter_clock(i + 2)
                    check_d = 1
            else:
                break
                
ans = 0
if gear[0][0] == 1:
    ans += 1
if gear[1][0] == 1:
    ans += 2
if gear[2][0] == 1:
    ans += 4
if gear[3][0] == 1:
    ans += 8
    
print(ans)
```
