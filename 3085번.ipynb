https://www.acmicpc.net/problem/3085

```python
n = int(input())
candy = []
for _ in range(n):
    colors = list(map(str, input()))
    candy.append(colors)
    
ans = 0

def width():
    global ans
    for k in range(n):
        countrow = 1
        for l in range(n-1):
            if candy[k][l] == candy[k][l+1]:
                countrow += 1
                ans = max(ans, countrow)
            else :
                countrow = 1
def height():
    global ans
    for k in range(n):
        countcolumn = 1
        for l in range(n-1):
            if candy[l][k] == candy[l+1][k]:
                countcolumn += 1
                ans = max(ans,countcolumn)
            else :
                countcolumn = 1
                
for i in range(n):
    for j in range(n - 1):
        if candy[i][j] != candy[i][j+1]:
            candy[i][j], candy[i][j+1] = candy[i][j+1], candy[i][j]
            width()
            height()
            candy[i][j+1], candy[i][j] = candy[i][j], candy[i][j+1]
        if candy[j][i] != candy[j+1][i]:
            candy[j][i], candy[j+1][i] = candy[j+1][i], candy[j][i]
            width()
            height()
            candy[j+1][i], candy[j][i] = candy[j][i], candy[j+1][i]
            
print(ans)
```
