```python
import sys
#input = sys.stdin.readline

t = int(input())
for _ in range(t):
    n = int(input())
    array = sorted([list(map(int, input().split())) for x in range(n)], key = lambda x : x[0])
    cnt = 1
    min = array[0][1]
    for i in range(1, n):
        if array[i][1] < min:
            min = array[i][1]
            cnt += 1
            
    print(cnt)
```
