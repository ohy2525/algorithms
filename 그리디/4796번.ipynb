https://www.acmicpc.net/problem/4796

```python
i = 1
while True:
    l, p, v = map(int, input().split())
    if l == 0 and p == 0 and v == 0:
        break
    ans = v // p * l
    if v % p >= l:
        ans += l
    else :
        ans += v % p
    print("Case {}: {}".format(i, ans))
    i += 1
```
