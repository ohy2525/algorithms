https://www.acmicpc.net/problem/1213

```python
import collections
word = input()
c_count = collections.Counter(word)

odd = 0
result = ''
mid = ''
for k, v in c_count.items():
    if v % 2 == 1:
        odd += 1
        mid = k
     
    if odd > 1:
        print("I'm Sorry Hansoo")
        exit()
             
for k, v in sorted(c_count.items()):
    result += k * (v // 2)
    
print(result + mid + result[::-1])
```
