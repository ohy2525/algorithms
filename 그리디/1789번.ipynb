https://www.acmicpc.net/problem/13305

```python
import sys
#input = sys.stdin.readline
n = int(input())
dist = list(map(int, input().split()))
cost = list(map(int, input().split()))

ans = dist[0] * cost[0]
c = cost[0]
for i in range(1, n-1):
    
    if cost[i] < c:
        c = cost[i]
        ans += dist[i] * c
    else:
        ans += dist[i] * c
    
print(ans)
```
