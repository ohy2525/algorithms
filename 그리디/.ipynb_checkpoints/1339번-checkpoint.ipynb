https://www.acmicpc.net/problem/1339

```python
n = int(input())
words = []
for _ in range(n):
    words.append(input())
dict = {}
for word in words:
    square = len(word) - 1
    for c in word:
        if c in dict:
            dict[c] += pow(10, square)
        else:
            dict[c] = pow(10, square)
        square -= 1
        
dict = sorted(dict.values(), reverse = True)
result = 0
num = 9
for value in dict:
    result += value * num
    num -= 1
print(result)
```
