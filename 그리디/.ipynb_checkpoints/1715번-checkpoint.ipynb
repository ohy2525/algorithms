https://www.acmicpc.net/problem/1715

```python
import sys
import heapq
#input = sys.stdin.readline
n = int(input())
arr = []
for _ in range(n):
    arr.append(int(input()))
heapq.heapify(arr)
ans = 0

while len(arr) > 1:
    tmp = heapq.heappop(arr) + heapq.heappop(arr)
    heapq.heappush(arr, tmp)
    ans += tmp
    
print(ans)
```
