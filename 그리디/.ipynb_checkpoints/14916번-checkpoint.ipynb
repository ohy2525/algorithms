```python
n = int(input())

answer = 0

while True:
    if n == 1 or n == 3:
        answer = -1
        break
    if n == 8:
        answer += 4
        break
    if n == 6:
        answer += 3
        break
        
    if n >= 5:
        n -= 5
        answer += 1
    else:
        answer += n // 2
        break

print(answer)
```

```python
n = int(input())

answer = 0

while True:
    if n % 5 == 0:
        answer += n // 5
        break
        
    else:
        n -= 2
        answer += 1
        
        if n < 0:
            answer = -1
            break
            
print(answer)
```
