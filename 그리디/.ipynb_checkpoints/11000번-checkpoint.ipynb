https://www.acmicpc.net/problem/11000

```python
import heapq

n = int(input())
lectures = [list(map(int, input().split())) for _ in range(n)]
lectures.sort()

ans = []
heapq.heappush(ans, lectures[0][1])

for i in range(1, n):
    if lectures[i][0] < ans[0]:
        heapq.heappush(ans, lectures[i][1])
    else:
        heapq.heappop(ans)
        heapq.heappush(ans, lectures[i][1])

print(len(ans))
```
