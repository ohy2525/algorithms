```python
def solution(m, n, board):
    answer = 0
    new_board = []
    for i in range(n):
        tmp = ''
        for j in reversed(range(m)):
            tmp += board[j][i]
        new_board.append(tmp)
      
    while True:
        num_list = set()
        visited = [[0] * m for _ in range(n)]
        for i in range(len(new_board)):
            for j in range(len(new_board[i])):
                check = new_board[i][j]
                if j + 1 < len(new_board[i]) and i + 1 < len(new_board):
                    if j + 1 < len(new_board[i + 1]):
                        if check == new_board[i][j + 1] == new_board[i + 1][j] == new_board[i + 1][j + 1]:
                            num_list.add((i, j))
                            num_list.add((i, j + 1))
                            num_list.add((i + 1, j))
                            num_list.add((i + 1, j + 1))

        if len(num_list) == 0:
            break
        answer += len(num_list)

        for num in num_list:
            visited[num[0]][num[1]] = 1

        for i in range(len(new_board)):
            tmp = ''
            for j in range(len(new_board[i])):
                if visited[i][j] == 0:
                    tmp += new_board[i][j]
            new_board[i] = tmp
        
    return answer
```

```python
m = 6
n = 6
board = ["AABBEE", "AAAEEE", "VAAEEV", "AABBEE", "AACCEE", "VVCCEE"]
solution(m, n, board)
```
