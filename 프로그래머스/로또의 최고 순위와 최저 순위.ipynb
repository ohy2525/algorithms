```python
def solution(lottos, win_nums):
    answer = []
    cnt = 0
    for i in lottos:
        if i in win_nums:
            cnt += 1
    cnt_0 = lottos.count(0)
    lotto = [6, 6, 5, 4, 3, 2, 1]
    
    answer.append(lotto[cnt + cnt_0])
    answer.append(lotto[cnt])
    
    return answer
```

```python
lottos = [0, 0, 0, 0, 0, 0]
win_nums = [38, 19, 20, 40, 15, 25]
solution(lottos, win_nums)
```

```python
lottos = [44, 1, 0, 0, 31, 25]
win_nums = [31, 10, 45, 1, 6, 19]
solution(lottos, win_nums)
```
