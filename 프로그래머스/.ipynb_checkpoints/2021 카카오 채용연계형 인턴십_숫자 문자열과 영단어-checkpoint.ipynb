```python
dic = {}
dic["zero"] = 0
dic["one"] = 1
dic["two"] = 2
dic["three"] = 3
dic["four"] = 4
dic["five"] = 5
dic["six"] = 6
dic["seven"] = 7
dic["eight"] = 8
dic["nine"] = 9

def solution(s):
    global dic
    answer = ""
    start = 0
    for i in range(1, len(s) + 1):
        tmp = s[start:i]
        try:
            num = int(tmp)
            answer += tmp
            start = i
        except:
            pass

        try:
            num = dic[tmp]
            answer += str(dic[tmp])
            start = i
        except:
            pass
        
    return int(answer)
```

```python
s = "one4seveneight"
solution(s)
```

```python
def solution(s):
    words = ['zero', 'one', 'two', 'three', 'four', 'five', 'six', 'seven', 'eight', 'nine']

    for i in range(len(words)):
        s = s.replace(words[i], str(i))

    return int(s)
```

```python
s = "one4seveneight"
solution(s)
```
