```python
n = 437674
k = 3
```

```python
def conversion(n, k):   
    tmp = ""
    while n:
        tmp += str(n % k)
        n = n // k
    return tmp[::-1]
```

```python
def prime_number(x):
    if x == 1:
        return False
    for i in range(2, int(x ** 0.5) + 1):
        if x % i == 0:
            return False
    return True
```

```python
def solution(n, k):
    num = conversion(n, k)
    answer = 0
    num_list = num.split("0")
    for num in num_list:
        if num and prime_number(int(num)):
            answer += 1
    return answer
```

```python
solution(n, k)
```
