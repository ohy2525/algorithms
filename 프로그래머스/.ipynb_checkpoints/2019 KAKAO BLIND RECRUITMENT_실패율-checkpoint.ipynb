```python
def solution(N, stages):
    answer = {}
    for i in range(1,N+1):
        try:
            tmp = len([x for x in stages if x == i])/len([x for x in stages if x >= i])
        except:
            tmp = 0
        answer[i] = tmp
    answer = sorted(answer, key = answer.get, reverse=True)
    return answer
```

```python
N = 5
stages = [2,1,2,6,2,4,3,3]
solution(N, stages)
```

```python
N = 4
stages = [4,4,4,4]
solution(N, stages)
```
