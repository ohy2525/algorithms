```python
phone_book = ["97674223", "119", "1195524421"]
```

```python
def solution(phone_book):
    phone_book.sort()   
    for i in range(len(phone_book) - 1):
        if phone_book[i + 1].startswith(phone_book[i]):
            return False
    return True
```

```python
solution(phone_book)
```

```python
def solution(phone_book):
    phone_book.sort()
    for p1, p2 in zip(phone_book, phone_book[1:]):
        if p2.startswith(p1):
            return False
    return True
```

```python
solution(phone_book)
```
