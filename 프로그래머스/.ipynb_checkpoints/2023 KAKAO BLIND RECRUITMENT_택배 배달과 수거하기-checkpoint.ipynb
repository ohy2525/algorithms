```python
def solution(cap, n, deliveries, pickups):
    answer = 0
    
    while deliveries or pickups:
        while deliveries and deliveries[-1] == 0:
            del deliveries[-1]
        while pickups and pickups[-1] == 0:
            del pickups[-1]
        answer += 2 * max(len(deliveries), len(pickups))
        
        can_del = cap
        for i in reversed(range(len(deliveries))):
            if can_del < deliveries[i]:
                deliveries[i] -= can_del
                break
            else:
                can_del -= deliveries[i]
                deliveries[i] = 0
        can_pick = cap
        for i in reversed(range(len(pickups))):
            if can_pick < pickups[i]:
                pickups[i] -= can_pick
                break
            else:
                can_pick -= pickups[i]
                pickups[i] = 0
                
    return answer
```

```python
cap = 4
n = 5
deliveries = [1, 0, 3, 1, 2]
pickups = [0, 3, 0, 4, 0]
solution(cap, n, deliveries, pickups)
```
