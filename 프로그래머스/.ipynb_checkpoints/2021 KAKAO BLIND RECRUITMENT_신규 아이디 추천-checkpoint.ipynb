```python
import re 

def solution(new_id):
    answer = ""
    new_id1 = new_id.lower()

    p = re.compile('[0-9a-zA-Z-_.]+')
    for i in p.findall(new_id1):
        answer += i

    answer = re.sub('[.]+', '.', answer)

    if answer[0] == "." and answer[-1] == ".":
        answer = answer[1:-1]
    elif answer[0] == ".":
        answer = answer[1:]
    elif answer[-1] == ".":
        answer = answer[:-1]

    if len(answer) == 0:
        answer = "a"

    answer = answer[:15]

    if answer[0] == "." and answer[-1] == ".":
        answer = answer[1:-1]
    elif answer[0] == ".":
        answer = answer[1:]
    elif answer[-1] == ".":
        answer = answer[:-1]

    while len(answer) < 3:
        answer += answer[-1]


    return answer
```

```python
new_id = "=.="
solution(new_id)
```

```python
import re

def solution(new_id):
    st = new_id
    st = st.lower()
    st = re.sub('[^a-z0-9\-_.]', '', st)
    st = re.sub('\.+', '.', st)
    st = re.sub('^[.]|[.]$', '', st)
    st = 'a' if len(st) == 0 else st[:15]
    st = re.sub('^[.]|[.]$', '', st)
    st = st if len(st) > 2 else st + "".join([st[-1] for i in range(3-len(st))])
    return st
```
