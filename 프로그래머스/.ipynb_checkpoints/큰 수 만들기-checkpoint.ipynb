```python
def solution(number, k):
    answer = []
    
    for i in number:
        while  k > 0 and len(answer) != 0 and answer[-1] < i:
            answer.pop()
            k -= 1
        answer.append(i)
        
    if k == 0:
        return ''.join(answer)
    else:
        return ''.join(answer[:-k])
```

```python
number = "4321"
k = 1
solution(number, k)
```
