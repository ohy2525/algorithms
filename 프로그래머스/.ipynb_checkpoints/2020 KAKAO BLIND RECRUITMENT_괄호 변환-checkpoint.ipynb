```python
from collections import deque
def separate(p):
    l, r = 0, 0
    for i in range(len(p)):
        if p[i] =='(':
            l += 1
        else:
            r += 1
        
        if l == r:
            return p[:i+1], p[i+1:]
```

```python
def check(u):
    q = deque()
    for i in u:
        if i == '(':
            q.append(i)
        else:
            if not q:
                return False
            q.pop()
    return True if not q else False
```

```python
def solution(p):
    if not p:
        return p
    u, v = separate(p)
    if check(u):
        return u + solution(v)
    else:
        answer = '('
        answer += solution(v)
        answer += ')'
        
        for i in u[1:len(u) - 1]: 
            if i == '(':
                answer += ')'
            else:
                answer += '('
    return answer 
```

```python
p = "()))((()"
solution(p)
```
