```python
def solution(n):
    answer = 0
    
    cnt = bin(n).count("1")
    
    for num in range(n + 1, 1000001):
        tmp = bin(num).count("1")
        
        if cnt == tmp:
            answer = num
            break
    
    return answer
```

```python
solution(78)
```
