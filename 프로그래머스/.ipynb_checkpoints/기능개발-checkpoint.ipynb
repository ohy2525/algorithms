```python
import math

def solution(progresses, speeds) :
    time = []
    answer = []
    for i in range(len(progresses)):
        time.append(int(math.ceil((100 - progresses[i]) / speeds[i])))

    now = time.pop(0)
    cnt = 1
    while len(time) > 0:
        if now >= time[0]:
            time.pop(0)
            cnt += 1
        else:
            answer.append(cnt)
            now = time.pop(0)
            cnt = 1
    answer.append(cnt) 
            
    return answer
```

```python
progresses = [93, 93]
speeds = [1, 1]
solution(progresses, speeds)
```
