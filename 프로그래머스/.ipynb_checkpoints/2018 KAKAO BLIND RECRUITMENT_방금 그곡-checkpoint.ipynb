```python
def clean(s):
    s = s.replace('C#', 'c')
    s = s.replace('D#', 'd')
    s = s.replace('F#', 'f')
    s = s.replace('G#', 'g')
    s = s.replace('A#', 'a')
    return s

def solution(m,musicinfos):
    answer=[0,'(None)']  
    m = clean(m)
    for info in musicinfos:
        split_info = info.split(',')
        time = (int(split_info[1][:2])-int(split_info[0][:2])) * 60 + int(split_info[1][-2:])-int(split_info[0][-2:])
        title = split_info[2]
        part_notes = clean(split_info[3])
        full_notes = part_notes * (time // len(part_notes)) + part_notes[:time%len(part_notes)]
        
        if m in full_notes and time > answer[0]:
            answer=[time, title]
            
    return answer[1]
```

```python
m = "ABCDEFG"
musicinfos = ["12:00,12:14,HELLO,CDEFGAB", "13:00,13:05,WORLD,ABCDEF"]
solution(m, musicinfos)
```

```python
m = "CC#BCC#BCC#BCC#B"
musicinfos = ["03:00,03:30,FOO,CC#B", "04:00,04:08,BAR,CC#BCC#BCC#B"]
solution(m, musicinfos)
```
