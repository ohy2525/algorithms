```python
def solution(fees, records):
    cars = set()
    for i in records:
        cars.add(i.split()[1])
    cars = list(cars)
    cars.sort()
    answer = [0] * len(cars)
    for i in range(len(cars)):
        car = cars[i]
        tmp = []
        for j in records:
            if j.split()[1] == car:
                tmp.append(j.split()[0])
        if len(tmp) % 2 == 1:
            tmp.append("23:59")
        t = 0
        for k in range(len(tmp) // 2):
            h1, m1 = tmp[2*k + 1].split(':')
            h2, m2 = tmp[2*k].split(':')
            m = int(m1) - int(m2)
            h = int(h1) - int(h2)
            t += 60 * h + m
        if t <= fees[0]:
            answer[i] += fees[1]
        else:
            if (t-fees[0]) % fees[2] != 0:
                fee = fees[1] + ((t - fees[0]) // fees[2] + 1) * fees[3]
            else:
                fee = fees[1] + ((t - fees[0]) // fees[2]) * fees[3]
            answer[i] += fee
    return answer
```

```python
fees = [180, 5000, 10, 600]
records = ["05:34 5961 IN", "06:00 0000 IN", "06:34 0000 OUT", "07:59 5961 OUT", "07:59 0148 IN", "18:59 0000 IN", "19:09 0148 OUT", "22:59 5961 IN", "23:00 5961 OUT"]
solution(fees, records)
```
