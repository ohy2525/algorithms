```python
from collections import deque

def bfs(x):
    visited = [False for _ in range(n + 1)]
    q = deque(x)
    visited[x] = True
    cnt = 1
    while q:
        s = q.popleft()
        for i in graph[s]:
            if not visited[i]:
                visited[i] = True
                q.append(i)
                cnt += 1
        return cnt
    
def solution(n, wires):
    answer = n
    graph = [[] for _ in range(n + 1)]
    for wire in wires:
        graph[wire[0]].append(wire[1])
        graph[wire[1]].append(wire[0])
        
    def bfs(x):
        visited = [False for _ in range(n + 1)]
        q = deque()
        q.append(x)
        visited[x] = True
        cnt = 1
        while q:
            s = q.popleft()
            for i in graph[s]:
                if not visited[i]:
                    visited[i] = True
                    q.append(i)
                    cnt += 1
            return cnt
        
    for wire in wires:
        graph[wire[0]].remove(wire[1])
        graph[wire[1]].remove(wire[0])
        
        answer = min(abs(bfs(wire[0]) - bfs(wire[1])), answer)
        
        graph[wire[0]].append(wire[1])
        graph[wire[1]].append(wire[0])
        
    return answer
```

```python
n = 7
wires = [[1,2],[2,7],[3,7],[3,4],[4,5],[6,7]]
solution(n, wires)
    
```
