```python
def solution(bridge_length, weight, truck_weights):
    answer = 0
    x = [0] * bridge_length
    s = 0

    while (x):
        answer += 1
        s -= x.pop(0)
        if truck_weights:
            if (s + truck_weights[0]) <= weight:
                s += truck_weights[0]
                x.append(truck_weights.pop(0)) 
            else:
                x.append(0)

    return answer
```

```python
bridge_length = 2
weight = 10
truck_weights = [7,4,5,6]
solution(bridge_length, weight, truck_weights)
```
