```python
def solution(today, terms, privacies):
    term_dict = {}
    for term in terms:
        a, b = term.split()
        term_dict[a] = int(b)
        
    y, m, d = today.split(".")
    today = int(d) + int(m) * 28 + int(y) * 28 * 12
    answer = []
    for i in range(len(privacies)):
        date, term = privacies[i].split()
        y, m, d = date.split(".")
        t = int(d) + int(m) * 28 + int(y) * 28 * 12 + term_dict[term] * 28
        if today >= t:
            answer.append(i + 1)

    return answer
```

```python
today = "2020.01.01"
terms = ["Z 12", "D 5"]
privacies = ["2019.01.01 Z", "2019.11.15 Z", "2019.08.02 D", "2019.07.01 D", "2018.12.28 Z"]

solution(today, terms, privacies)
```
