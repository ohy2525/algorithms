```python
brown = 24
yellow = 24
```

```python
def solution(brown, yellow):
    answer = []
    for i in range(1, 2501):
        for j in range(1, 2501):
            if i * j == brown + yellow and (i + j) * 2 == brown + 4:
                answer.append(max(i, j))
                answer.append(min(i, j))
                break
        else:
            continue
        break
    return answer
```

```python
solution(brown, yellow)
```
