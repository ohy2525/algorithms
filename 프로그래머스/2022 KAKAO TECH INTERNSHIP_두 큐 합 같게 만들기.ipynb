```python
queue1 = [1,1]
queue2 = [1,5]
```

```python
from collections import deque


def solution(queue1, queue2):
    q1, q2 = deque(queue1), deque(queue2)
    sum1, sum2 = sum(q1), sum(q2)
    if (sum1 + sum2) % 2 != 0:
        return -1

    answer = 0
    while True:
        if answer == 3 * len(queue1):
            return -1

        if sum1 == sum2:
            return answer
        elif sum1 > sum2:
            tmp = q1.popleft()
            q2.append(tmp)
            sum1 -= tmp
            sum2 += tmp
        else:
            tmp = q2.popleft()
            q1.append(tmp)
            sum1 += tmp
            sum2 -= tmp
        answer += 1
    
```

```python
solution(queue1, queue2)
```

```python
def solution(queue1, queue2):
    target = (sum(queue1) + sum(queue2)) // 2
    cur = sum(queue1)
    queue3 = queue1 + queue2 + queue1

    s = 0
    e = len(queue1) - 1
    answer = 0
    while True:
        if cur == target:
            return answer
        if cur < target:
            e += 1
            if e >= len(queue3):
                return -1
            cur += queue3[e]
        else:
            cur -= queue3[s]
            s += 1
        answer += 1
```

```python
solution(queue1, queue2)
```
