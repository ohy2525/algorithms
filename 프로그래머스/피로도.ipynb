```python
k = 80
dungeons = [[80,20],[50,40],[30,10]]
```

```python
import itertools

def solution(k, dungeons):
    p = itertools.permutations(dungeons, len(dungeons))
    answer = 0
    for i in list(p):
        hp = k
        ans = 0
        for j in range(len(dungeons)):
            if i[j][0] <= hp:
                hp -= i[j][1]
                ans += 1
                if hp < 0:
                    break
        if ans > answer:
            answer = ans
    return answer
```

```python
solution(k, dungeons)
```
