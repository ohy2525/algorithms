```python
def solution(numlist, n):
    arr = []
    for num in numlist:
        arr.append([abs(num - n), num])
    answer = []

    sorted_list = sorted(arr, key = lambda x: (x[0], -x[1]))

    for tmp in sorted_list:
        answer.append(tmp[1])
        
    return answer
```

```python
numlist = [1,2,3,4,5,6]
n = 4
solution(numlist, n)
```
