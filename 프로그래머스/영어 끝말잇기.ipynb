```python
n = 5
words = ["hello", "observe", "effect", "take", "either", "recognize", "encourage", "ensure", "establish", "hang", "gather", "refer", "reference", "estimate", "executive"]
```

```python
def solution(n, words):
    answer = 0
    tmp = [words[0]]
    result = [0, 0]
    for i in range(1, len(words)):
        if words[i][0] != words[i - 1][-1]:
            answer = i
            break

        else:
            if words[i] in tmp:
                answer = i
                break
            else:
                tmp.append(words[i])

    if answer == 0:
         return result
        
    result[0] = (answer + 1) % n if (answer + 1) % n != 0 else n
    result[1] = answer // n + 1

    return result
```

```python
solution(n, words)
```
