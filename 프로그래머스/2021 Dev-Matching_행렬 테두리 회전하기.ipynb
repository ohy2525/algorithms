```python
from collections import deque

def solution(rows, columns, queries):
    arr = [[i+columns*j for i in range(1,columns+1)] for j in range(rows)]
    q = deque()
    answer = []
    
    for i in queries:
        x1, y1, x2, y2 = i[0]-1,i[1]-1,i[2]-1,i[3]-1
        a,b,c,d = i[0]-1,i[1]-1,i[2]-1,i[3]-1
        
        for i in range(y2 - y1):
            q.append(arr[x1][y1 + i])
            
        for j in range(x2 - x1):
            q.append(arr[x1 + j][y2])
            
        for k in range(y2 - y1):
            q.append(arr[x2][y2 - k])
            
        for l in range(x2 - x1):
            q.append(arr[x2-l][y1])
            
        q.rotate(1)
        answer.append(min(q))
        
        for i in range(y2 - y1):
            arr[x1][y1 + i] = q[0]
            q.popleft()
        for j in range(x2 - x1):
            arr[x1 + j][y2] = q[0]
            q.popleft()
        for k in range(y2 - y1):
            arr[x2][y2 - k] = q[0]
            q.popleft()
        for l in range(x2 - x1):
            arr[x2 - l][y1] = q[0]
            q.popleft()
    return answer
```

```python
rows = 100
columns = 97
queries = [[1,1,100,97]]
```

```python
solution(rows, columns, queries)
```
