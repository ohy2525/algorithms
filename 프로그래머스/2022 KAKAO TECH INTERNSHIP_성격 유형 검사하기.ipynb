```python
survey = ["AN", "CF", "MJ", "RT", "NA"]
choices = [5, 3, 2, 7, 5]
```

```python
def solution(survey, choices):
    answer = ''
    test = {"A":0, "N":0, "C":0, "F":0, "M":0, "J":0, "R":0, "T":0,}
    for i in range(len(survey)):
        type1, type2 = survey[i][0], survey[i][1]
        if choices[i] == 1:
            test[type1] += 3
        elif choices[i] == 2:
            test[type1] += 2
        elif choices[i] == 3:
            test[type1] += 1
        elif choices[i] == 5:
            test[type2] += 1
        elif choices[i] == 6:
            test[type2] += 2
        elif choices[i] == 7:
            test[type2] += 3
    if test["R"] >= test["T"]:
        answer += "R"
    else:
        answer += "T"

    if test["C"] >= test["F"]:
        answer += "C"
    else:
        answer += "F"
    if test["J"] >= test["M"]:
        answer += "J"
    else:
        answer += "M"
    if test["A"] >= test["N"]:
        answer += "A"
    else:
        answer += "N"
        
    return answer
```
