```python
def solution(m, n, puddles):
    dp = [([0] * m) for _ in range(n)]
    arr = [([True] * m) for _ in range(n)]

    num = 1000000007
    for puddle in puddles:
        arr[puddle[1] - 1][puddle[0] - 1] = False

    dp[0][0] = 1
    for i in range(n):
        for j in range(m):
            if  j + 1 < m and arr[i][j + 1]:
                dp[i][j + 1] = dp[i][j + 1] + dp[i][j] % num
            if i + 1 < n and arr[i + 1][j]:
                dp[i + 1][j] = dp[i + 1][j] + dp[i][j] % num
                        
    print(dp)
    return dp[n - 1][m - 1]
```

```python
def solution(m, n, puddles):
    dp = [[0] * (m + 1) for _ in range(n + 1)]  
    dp[1][1] = 1
    for i in range(1, n + 1):
        for j in range(1, m + 1):
            if [j,i] in puddles or [i, j] == [1, 1]:
                continue
            else:
                dp[i][j] = dp[i - 1][j] + dp[i][j - 1]
                
    return (dp[-1][-1] % 1000000007)
```

```python
m = 4
n = 3
puddles = [[2, 1], [2,2]]
solution(m, n, puddles)
```
