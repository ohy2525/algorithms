```python
X = "5525"
Y = "1255"
```

```python
def solution(X, Y):
    answer = []
    for i in (set(X) & set(Y)) :
        for j in range(min(X.count(i), Y.count(i))) :
            answer.append(i)
            
    answer.sort(reverse=True)
    
    if len(answer) == 0:
        return "-1"
    
    if answer[0] == "0":
        return "0"
    
    answer = "".join(answer)
    
    return answer
```

```python
X = "5525"
Y = "1255"
```

```python
solution(X, Y)
```

```python
def solution(X, Y):
    answer = ''

    for i in range(9, -1, -1) :
        answer += (str(i) * min(X.count(str(i)), Y.count(str(i))))

    if answer == '' :
        return '-1'
    elif len(answer) == answer.count('0'):
        return '0'
    else :
        return answer
```

```python
solution(X, Y)
```
