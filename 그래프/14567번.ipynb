https://www.acmicpc.net/problem/14567

```python
import sys
#input = sys.stdin.readline
n, m = map(int, input().split())

graph = [1] * (n + 1)  
array = [[] for _ in range(n + 1)]
for i in range(m):
    a, b = map(int, input().split())
    array[a].append(b)

for i in range(1, n + 1):
    for j in array[i]:
        graph[j] = max(graph[j], graph[i] + 1)
print(*graph[1:])
```
