```python
import sys
#input = sys.stdin.readline
n = int(input())
graph = []
for i in range(n):
    graph.append(list(input().strip()))
    
visited = [[0] * n for i in range(n)]

for k in range(n):
    for i in range(n):
        for j in range(n):
            if i == j:
                continue
            if graph[i][j] == "Y" or (graph[i][k] == "Y" and graph[k][j] == "Y"):
                visited[i][j] = 1
ans = 0
for i in range(n):
    cnt = 0
    for j in range(n):
        if visited[i][j] == 1:
            cnt += 1
    ans = max(ans, cnt)

print(ans)
```
