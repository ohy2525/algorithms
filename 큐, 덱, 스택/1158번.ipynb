https://www.acmicpc.net/problem/1158

```python
from collections import deque
n, k = map(int, input().split())
q = deque()
for i in range(1, n +1):
    q.append(i)
res = []
while q:
    q.rotate(-(k-1))
    res.append(q.popleft())
print('<', end = '')
print(', '.join(map(str, res)), end = '')
print('>')
```
