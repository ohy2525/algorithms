https://www.acmicpc.net/problem/2579

```python
n = int(input())
array = []
dp = [0] * 301
for _ in range(n):
    array.append(int(input()))
dp[0] = array[0]
dp[1] = array[0] + array[1]
dp[2] = max(array[0] + array[2], array[1] + array[2])
for i in range(3, n):
    dp[i] = max(dp[i-2] + array[i], dp[i-3] + array[i-1] + array[i])
    
print(dp[n-1])
```
