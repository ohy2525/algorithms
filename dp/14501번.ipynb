https://www.acmicpc.net/problem/14501

```python
n = int(input())
t = []
p = []
for _ in range(n):
    a, b = map(int, input().split())
    t.append(a)
    p.append(b)
    
dp = [0] * n  

for i in range(n):
    k = i + t[i] - 1  # 
    if k >= n: #   
        continue
    if i == 0: #out of index   
        dp[k] = p[0]
    else:      #      +   
        dp[k] = max(dp[k], max(dp[:i]) + p[i])

print(max(dp))
```
