```python
n = int(input())
arr = list(map(int, input().split()))

def solve(mapping: dict) -> int:
    result, cnt = 0, 0
    for x in arr:
        cnt += mapping[x]
        cnt = max(0, cnt)
        result = max(result, cnt)
    return result

print(max(solve({1:1, 2:-1}), solve({1:-1, 2:1})))
```

```python
n = int(input())
arr = list(map(int, input().split()))

arr_1 = list(map(lambda x : 1 if x == 1 else -1, arr))
arr_2 = list(map(lambda x : 1 if x == 2 else -1, arr))

ans_1 = [arr_1[0]]
ans_2 = [arr_2[0]]

for i in range(n - 1):
    ans_1.append(max(ans_1[i] + arr_1[i + 1], arr_1[i + 1]))
for i in range(n - 1):
    ans_2.append(max(ans_2[i] + arr_2[i + 1], arr_2[i + 1]))
    
print(max(max(ans_1), max(ans_2)))
```
