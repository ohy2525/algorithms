https://www.acmicpc.net/problem/1890

```python
import sys
#input = sys.stdin.readline
n = int(input())
graph = []
for _ in range(n):
    graph.append(list(map(int, input().split())))
    
cnt = [[0] * n for _ in range(n)]
cnt[0][0] = 1
        
for i in range(n):
    for j in range(n):
        if i == n - 1 and j == n - 1:
            break
        if cnt[i][j] != 0:
            if i + graph[i][j] < n:
                cnt[i + graph[i][j]][j] += cnt[i][j]
            if j + graph[i][j] < n:
                cnt[i][j + graph[i][j]] += cnt[i][j]
            
print(cnt[n - 1][n - 1])
```
