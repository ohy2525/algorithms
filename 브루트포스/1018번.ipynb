https://www.acmicpc.net/problem/1018

```python
m, n = map(int, input().split())
chess = []
for _ in range(m):
    chess.append(input())
    
```

```python
ans = []
for i in range(m-7):
    for j in range(n-7):
        first_W = 0
        first_B = 0
        for h in range(i, i+8):
            for w in range(j, j+8):
                if (h+w)%2 == 0:
                    if chess[h][w] != 'W':
                        first_W += 1
                    else:
                        first_B += 1
                else :
                    if chess[h][w] != 'B':
                        first_W += 1
                    else :
                        first_B += 1
        ans.append(min(first_W, first_B))    
                       
        
print(min(ans))  
```
