```python
def self_number(x):
    num = str(x)
    list_ = []
    for i in range(len(num)):
        list_.append(int(num[i]))
        
    return sum(list_) + x

a = [0]* 10000
for i in range(1, 10001):
    a[i-1] = i
    
b = []
for i in range(1, 10001):
    b.append(self_number(i))
    
c = set(a) - set(b)
d = list(c)
d.sort()
for i in range(len(d)):
    print(d[i])
```

```python
num = set(range(1, 10001))
new_num = set()

for i in range(1, 10001):
    for j in str(i):
        i += int(j)
    new_num.add(i)
    
self_num = sorted(num - new_num)
for i in self_num:
    print(i)
```
