```python
from collections import deque

def bfs():
    queue = deque()
    queue.append(n)
    while queue:
        x = queue.popleft()
        if x == k:
            print(visited[x])
            break
        for nx in (x - 1, x + 1, x * 2):
            if 0 <= nx < 100001 and visited[nx] == 0:
                queue.append(nx)
                visited[nx] = visited[x] + 1
                
visited = [0] * 100001
n, k = map(int, input().split())
bfs()
```
