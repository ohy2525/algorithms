```python
from collections import deque
import copy

direction = [(-1, 0), (1, 0), (0, -1), (0, 1)]

def rotate(x, y, angle):
    new_graph = copy.deepcopy(graph)
    q = deque([])
    for i in range(y - 1, y + 2):
        q.append(new_graph[x- 1][i])
    for i in range(x - 1, x + 2):
        q.append(new_graph[i][y + 1])
    for i in range(y + 1, y - 2, -1):
        q.append(new_graph[x + 1][i])
    for i in range(x + 1, x - 2, -1):
        q.append(new_graph[i][y - 1])
    
    if angle == 0:
        q.rotate(3)
    elif angle == 1:
        q.rotate(6)
    else:
        q.rotate(9)

    for i in range(y -1, y + 2):
        new_graph[x - 1][i] = q.popleft()
    for i in range(x - 1, x + 2):
        new_graph[i][y + 1] = q.popleft()
    for i in range(y + 1, y - 2, -1):
        new_graph[x + 1][i] = q.popleft()
    for i in range(x + 1, x - 2, -1):
        new_graph[i][y - 1] = q.popleft()

    return new_graph


def dfs(x, y):
    global cnt, visited, new_graph, tmp_cnt, tmp_list
    tmp_cnt += 1
    tmp_list.append((x, y))
    target = new_graph[x][y]
    for (dx, dy) in direction:
        nx = x + dx
        ny = y + dy
        if 0 <= nx < 5 and 0 <= ny < 5 and not visited[nx][ny] and new_graph[nx][ny] == target:
            visited[nx][ny] = True
            dfs(nx, ny)




k, m = map(int, input().split())
graph = []
for _ in range(5):
    graph.append(list(map(int, input().split())))
arr = list(map(int, input().split()))
idx = 0
beforeIdx = 0
answer = []
for _ in range(k):
    res = []

    for i in range(1, 4):
        for j in range(1, 4):
            for angle in range(3):
                new_graph = rotate(i, j, angle)
                cnt = 0
                visited = [[False] * 5 for _ in range(5)]
                for x in range(5):
                    for y in range(5):
                        if not visited[x][y]:
                            visited[x][y] = True
                            tmp_cnt = 0
                            tmp_list = []
                            dfs(x, y)
                            if tmp_cnt >= 3:
                                cnt += tmp_cnt
                                for (a, b) in tmp_list:
                                    new_graph[a][b] = 0

                res.append((i, j, cnt, angle, new_graph))

    res.sort(key = lambda x : (-x[2], x[3], x[1], x[1]))

    if res[0][2] == 0:
        break
    graph = res[0][4]
    while True:
        for j in range(5):
            for i in range(4, -1, -1):
                if graph[i][j] == 0:
                    graph[i][j] = arr[idx]
                    idx += 1
        cnt = 0
        visited = [[False] * 5 for _ in range(5)]
        for x in range(5):
            for y in range(5):
                if not visited[x][y]:
                    visited[x][y] = True
                    tmp_cnt = 0
                    tmp_list = []
                    new_graph = graph
                    dfs(x, y)
                    if tmp_cnt >= 3:
                        cnt += tmp_cnt
                        for (a, b) in tmp_list:
                            new_graph[a][b] = 0
        graph = new_graph
        if cnt == 0:
            break
    
    answer.append(idx - beforeIdx)
    beforeIdx = idx

print(*answer)
```
