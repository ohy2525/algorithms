```python
from collections import deque

laser_d = [(0, 1), (1, 0), (0, -1), (-1, 0)]
shell_d = [(-1, -1), (-1, 0), (-1, 1), (0, -1), (0, 0), (0, 1), (1, -1), (1, 0), (1, 1)]

def find_attack():
    attack_list = []
    for i in range(n):
        for j in range(m):
            if graph[i][j] > 0:
                attack_list.append((i, j, graph[i][j], history[i][j], (i + j), j))

    attack_list.sort(key = lambda x : (x[2], -x[3], -x[4], -x[5]))
    return attack_list[0][0], attack_list[0][1]

def find_target():
    target_list = []
    for i in range(n):
        for j in range(m):
            if graph[i][j] > 0:
                target_list.append((i, j, graph[i][j], history[i][j], (i + j), j))

    target_list.sort(key = lambda x : (-x[2], x[3], x[4], x[5]))
    return target_list[0][0], target_list[0][1]

def bfs():
    queue = deque([(attack_x, attack_y, [])])
    visited = [[False] * m for _ in range(n)]
    visited[attack_x][attack_y] = True

    while queue:
        x, y, path = queue.popleft()
        if (x, y) == (target_x, target_y):
            return True, path

        for dx, dy in laser_d:
            nx = (x + dx) % n
            ny = (y + dy) % m 
            if graph[nx][ny] > 0 and not visited[nx][ny]:
                npath = path + [(nx, ny)]
                visited[nx][ny] = True
                queue.append((nx, ny, npath))

    return False, path

n, m, k = map(int, input().split())
graph = []
for _ in range(n):
    graph.append(list(map(int, input().split())))

answer = 0
history = [[0] * m for _ in range(n)]
for t in range(k):
    attack_x, attack_y = find_attack()
    target_x, target_y = find_target()
    history[attack_x][attack_y] = t + 1
    graph[attack_x][attack_y] += (n + m)

    res, path = bfs()
    
    if not res:
        path = []
        for (dx, dy) in shell_d:
            nx = (target_x + dx) % n
            ny = (target_y + dy) % m
            if graph[nx][ny] > 0 and (nx, ny) != (attack_x, attack_y):
                path.append((nx, ny))

    for (x, y) in path:
        if x == target_x and y == target_y:
            graph[x][y] -= graph[attack_x][attack_y]
        else:
            graph[x][y] -= (graph[attack_x][attack_y] // 2)

    for i in range(n):
        for j in range(m):
            if graph[i][j] > 0 and (i, j) != (attack_x, attack_y) and (i, j) not in path:
                graph[i][j] += 1
    cnt = 0
    for i in range(n):
        for j in range(m):
            if graph[i][j] > 0:
                cnt += 1
    if cnt == 1:
        break

for g in graph:
    answer = max(answer, max(g))


print(answer)
```
