```python
from collections import deque

MAX_L = 70
R, C, K = 0, 0, 0
graph = [[0] * MAX_L for _ in range(MAX_L + 3)]
dx = [-1, 0, 1, 0]
dy = [0, 1, 0, -1]
isExit = [[False] * MAX_L for _ in range(MAX_L + 3)]
answer = 0

def inRange(x, y):
    return 3 <= x < R + 3 and 0 <= y < C

def resetMap():
    for i in range(R + 3):
        for j in range(C):
            graph[i][j] = 0
            isExit[i][j] = False

def canGo(x, y):
    flag = 0 <= y - 1 and y + 1 < C and x + 1 < R + 3
    flag = flag and (graph[x - 1][y - 1] == 0)
    flag = flag and (graph[x - 1][y] == 0)
    flag = flag and (graph[x - 1][y + 1] == 0)
    flag = flag and (graph[x][y - 1] == 0)
    flag = flag and (graph[x][y] == 0)
    flag = flag and (graph[x][y + 1] == 0)
    flag = flag and(graph[x + 1][y] == 0)
    return flag

def bfs(x, y):
    result = x
    q = deque([(x, y)])
    visited = [[False] * C for _ in range(R + 3)]
    visited[x][y] = True
    while q:
        cur_x, cur_y = q.popleft()
        for i in range(4):
            nx, ny = cur_x + dx[i], cur_y + dy[i]
            if inRange(nx, ny) and not visited[nx][ny] and (graph[nx][ny] == graph[cur_x][cur_y] or graph[nx][ny] != 0 and isExit[cur_x][cur_y]):
                q.append((nx, ny))
                visited[nx][ny] = True
                result = max(result, nx)
    return result

def down(x, y, d, id):
    if canGo(x + 1, y):
        down(x + 1, y, d, id)
    elif canGo(x + 1, y - 1):
        down(x + 1, y - 1, (d - 1) % 4, id)
    elif canGo(x + 1, y + 1):
        down(x + 1, y + 1, (d + 1) % 4, id)
    else:
        if not inRange(x - 1, y - 1) or not inRange(x + 1, y + 1):
            resetMap()
        else:
            graph[x][y] = id
            for i in range(4):
                graph[x + dx[i]][y + dy[i]] = id
            isExit[x + dx[d]][y + dy[d]] = True
            global answer
            answer += bfs(x, y) - 3 + 1
    
def main():
    global R, C, K
    R, C, K = map(int, input().split())
    for id in range(1, K + 1):
        c, d = map(int, input().split())
        down(0, c - 1, d, id)
    print(answer)
    
if __name__ == "__main__":
    main()
```
