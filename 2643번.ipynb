https://www.acmicpc.net/problem/2643

```python
n = int(input())
paper = []
d = [1] * n
for _ in range(n):
    w, h = map(int, input().split())
    paper.append((max(w, h), min(w, h)))
paper.sort(reverse = True)
                 
for i, length in enumerate(paper):
    for j in range(i):
        if min(paper[j][0], paper[j][1]) >= min(length[0], length[1]):
            d[i] = max(d[j] + 1, d[i])
            
print(max(d))
```
