```python
from collections import defaultdict
```

```python
n = int(input())
rank = list(map(int, input().split()))
graph = defaultdict(list)
indegree = defaultdict(int)
```

```python
graph = defaultdict(list)
indegree = defaultdict(int)
```

```python
for i in range(n-1):
    print('i = ', i)
    graph[rank[i]] = rank[i+1:]
    print('graph = ', graph)
    indegree[rank[i+1]] = i+1
    print('indegree = ', indegree)
```

```python
m = int(input())
for _ in range(m):
    a, b = map(int, input().split())
    
    if b in graph[a]:
        graph[b].append(a)
        graph[a].remove(b)
        
        indegree[a] += 1
        indegree[b] -= 1
    else :
        graph[a].append(b)
        graph[b].remove(a)
        indegree[a] -= 1
        indegree[b] += 1
```

```python
p = []
for c in rank:
    if indegree[c] == 0:
        p.append(c)
        
ans = []

while p:
    cur = p.pop()
    ans.append(cur)
    
    for i in graph[cur]:
        indegree[i] -=1
        
        if indegree[i] == 0:
            p.append(i)
            
if len(ans) == n:
    print(*ans)
else :
    print('IMPOSSIBLE')
```
