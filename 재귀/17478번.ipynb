```python
def func(m):
    print("_" * (4 * (n - m)) + '" ?"')

    if not m:
        print("_" * (4 * (n - m)) + '"    "')
        print("_" * (4 * (n - m)) + " .")
        return

    print("_" * (4 * (n - m)) + '" .          .')
    print("_" * (4 * (n - m)) + "       ,    .")
    print("_" * (4 * (n - m)) + '    .   ,      ."')
    func(m - 1)
    print("_" * (4 * (n - m)) + " .")


n = int(input())
print("       .")
func(n)
```
