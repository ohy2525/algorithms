```python
n = int(input())
paper = [list(map(int, input().split())) for _ in range(n)]
minus = 0
zero = 0
plus = 0
```

```python
def solution(x, y, n):
    global minus, zero, plus
    num = paper[x][y]
    for i in range(x, x + n):
        for j in range(y, y + n):
            if num != paper[i][j]:
                n = n // 3
                solution(x, y, n)
                solution(x, y + n, n)
                solution(x, y + n*2, n)
                solution(x + n, y, n)
                solution(x + n, y + n, n)
                solution(x + n, y + n*2, n)
                solution(x + n*2, y, n)
                solution(x + n*2, y + n, n)
                solution(x + n*2, y + n*2, n)
                return
            
    if num == -1:
        minus += 1
    elif num == 0:
        zero += 1
    elif num == 1:
        plus += 1
solution(0,0,n)
print(minus)
print(zero)
print(plus)
```
