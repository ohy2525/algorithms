https://www.acmicpc.net/problem/1992

```python
n = int(input())
video = [list(map(int, input())) for _ in range(n)]
ans = []

def solution(x, y, n):
    type = video[x][y]
    for i in range(x, x + n):
        for j in range(y, y + n):
            if type != video[i][j]:
                ans.append('(')
                solution(x, y, n//2)
                solution(x, y+n//2, n//2)
                solution(x+n//2, y, n//2)
                solution(x+n//2, y+n//2, n//2)
                ans.append(')')
                return
    
    if type == 0:
        ans.append(0)
    else :
        ans.append(1)
        
solution(0,0,n)

for k in ans:
    print(k, end='')
```
