https://www.acmicpc.net/problem/2644

```python
import sys
sys.setrecursionlimit(100000)

n = int(input())
x, y = map(int, input().split())
m = int(input())
graph = [[] for _ in range(n+1)]
visited = [False] * (n+1)
result = []

for _ in range(m):
    a, b = map(int, input().split())  
    graph[a].append(b)
    graph[b].append(a)


def dfs(x, num):
    visited[x] = True
    if x == y:
        result.append(num)
        
    num += 1
    for i in graph[x]:
        if not visited[i]:
            dfs(i, num)

dfs(x, 0)
if len(result) == 0:
    print(-1)
else:
    print(result[0])
```
