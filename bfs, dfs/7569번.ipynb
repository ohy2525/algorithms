```python
from collections import deque
import sys
#input = sys.stdin.readline
m, n, h = map(int, input().split())
graph = [[] for _ in range(h)]
for z in range(h):
    for _ in range(n):
        graph[z].append(list(map(int, input().split())))
        
queue = deque()
for z in range(h):
    for y in range(n):
        for x in range(m):
            if graph[z][y][x] == 1:
                queue.append((z, y, x))

dx = [1, -1, 0, 0, 0, 0]
dy = [0, 0, -1, 1, 0, 0]
dz = [0, 0, 0, 0, 1, -1]

def dfs():
    while queue:
        z, y, x = queue.popleft()
        for i in range(6):
            nx = x + dx[i]
            ny = y + dy[i]
            nz = z + dz[i]

            if 0 <= nx < m and 0 <= ny < n and 0 <= nz < h and graph[nz][ny][nx] == 0:
                graph[nz][ny][nx] = graph[z][y][x] + 1
                queue.append((nz, ny, nx))

dfs()

ans = 0
no = 0
for z in graph:
    for y in z:
        for x in y:
            if x == 0:
                no = 1
        ans = max(max(y), ans)
if no == 1:
    print(-1)
else :    
    print(ans -1)
```
