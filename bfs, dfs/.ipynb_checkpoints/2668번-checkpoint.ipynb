https://www.acmicpc.net/problem/2668

```python
n = int(input())
graph = [[] for _ in range(n + 1)]
for i in range(n):
    graph[i + 1].append(int(input()))
    
def dfs(num):
    if visited[num] == False:
        visited[num] = True
        for x in graph[num]:
            tmp_1.add(num)
            tmp_2.add(x)
            if tmp_1 == tmp_2:
                ans.extend(list(tmp_2))
                return
            dfs(x)
            
ans = []
for i in range(1, n + 1):
    visited = [False] * (n + 1)
    tmp_1 = set()
    tmp_2 = set()
    dfs(i)
    
answer = list(set(ans))
answer.sort()
print(len(answer))
for a in answer:
    print(a)
```
