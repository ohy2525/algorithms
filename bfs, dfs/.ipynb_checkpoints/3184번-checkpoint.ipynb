```python
from collections import deque
#input = sys.stdin.readline
dx = [-1, 1, 0, 0]
dy = [0, 0, -1, 1]

def bfs(x, y):
    global o, v
    cnt_o, cnt_v = 0, 0
    queue = deque()
    queue.append((x, y))
    if graph[x][y] == "v":
        cnt_v += 1
    elif graph[x][y] =="o":
        cnt_o += 1
        
    while queue:
        x, y = queue.popleft()
        for i in range(4):
            nx = x + dx[i]
            ny = y + dy[i]
            if 0 <= nx < r and 0 <= ny < c and visited[nx][ny] == False and graph[nx][ny] != "#":
                if graph[nx][ny] == "o":
                    cnt_o += 1
                elif graph[nx][ny] == "v":
                    cnt_v += 1
                    
                visited[nx][ny] = True
                queue.append((nx, ny))
   
    if cnt_o > cnt_v :
        v -= cnt_v
    else:
        o -= cnt_o
        

r, c = map(int, input().split())
graph = [list(input()) for _ in range(r)]
visited = [[False] * c for _ in range(r)]

o, v = 0, 0

for i in graph:
    for j in i:
        if j == "o":
            o += 1
        if j == "v":
            v += 1
            
for i in range(r):
    for j in range(c):
        if (graph[i][j] == "o" or graph[i][j] == "v") and visited[i][j] == False:
            visited[i][j] = True
            bfs(i, j)
            
print(o, v)
```
