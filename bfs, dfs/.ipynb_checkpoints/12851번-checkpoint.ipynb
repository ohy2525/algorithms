https://www.acmicpc.net/problem/12851

```python
from collections import deque

n, k = map(int, input().split())
visited = [0] * 100001
cnt = 0
def bfs(n):
    q = deque()
    q.append(n)
    global cnt
    while q:
        x = q.popleft()
        if x == k:
            cnt += 1
            continue
            
        for nx in (x - 1, x + 1, 2 * x):
            if 0 <= nx <= 100000:
                if visited[nx] == 0 or visited[nx] == visited[x] + 1:
                    q.append(nx)
                    visited[nx] = visited[x] + 1
                    
bfs(n)

print(visited[k])
print(cnt)
```
