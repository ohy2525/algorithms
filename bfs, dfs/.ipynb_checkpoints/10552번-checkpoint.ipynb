https://www.acmicpc.net/problem/10552

한 노인 주택에서 N명의 어르신들이 TV를 시청하고 있습니다. TV 프로그램은 1부터 M까지의 숫자로 표시되는 M개의 채널로 구성됩니다. 각각의 노인들은 자신이 좋아하는 채널과 싫어하는 채널이 있습니다.

만약 TV가 어떤 노인의 싫어하는 채널을 방영하고 있다면, 그 노인은 일어나 천천히 TV로 걸어가 해당 채널을 자신이 좋아하는 채널로 바꾸고 다시 편안한 의자에 앉습니다. 만약 현재 방영되고 있는 채널을 싫어하는 노인이 여러 명이라면, 그 중 가장 어린 노인이 일어나 채널을 바꾸고 나머지는 그 자리에 앉아 있습니다.

한 번의 채널 변경 이후에도 다른 노인이 새로운 채널을 싫어해서 계속해서 채널 변경이 일어날 수 있습니다. 노인들은 고집스러우므로 이런 일이 계속해서 일어날 수 있습니다.

노인들의 좋아하는 채널과 싫어하는 채널, 그리고 TV에서 초기에 방영되는 채널이 주어졌을 때, 노인들이 행복하고 앉아 있는 상태로 유지되기 위해 몇 번의 채널 변경이 필요한지 계산합니다.

입력의 첫 번째 줄에는 세 개의 정수 N, M 및 P (1 ≤ N, M ≤ 105, 1 ≤ P ≤ M)이 포함되어 있으며, 노인들의 수, TV 채널 수 및 TV에서 초기에 표시되는 채널이 포함됩니다.

이어지는 각각의 N개의 줄에는 각각의 노인의 좋아하는 채널과 싫어하는 채널을 나타내는 두 개의 정수 ai 및 bi (1 ≤ ai, bi ≤ M, ai ≠ bi)이 포함됩니다.

입력에서 노인들의 순서는 나이가 어린 순서부터입니다.

출력은 첫 번째 줄에 요구되는 채널 변경의 수를 포함해야 합니다. 채널 변경이 계속해서 무한정 일어날 경우, -1을 출력합니다.

```python
from collections import deque
import sys
#input = sys.stdin.readline
n, m, p = map(int, input().split())

graph = [0 for i in range(m + 1)]
visited = [False for i in range(m + 1)]
for i in range(n):
    a, b = map(int, input().split())
    if graph[b] == 0:
        graph[b] = a
        
def dfs(start):
    q = deque()
    q.append(start)
    visited[start] = True
    cnt = 0
    while q:
        x = q.popleft()
        if visited[graph[x]]:
            return -1
        
        if graph[x] != 0 and not visited[graph[x]]:
            q.append(graph[x])
            visited[graph[x]] = True
            cnt += 1
        
        if graph[x] == 0:
            break
            
    return cnt

print(dfs(p))
```
