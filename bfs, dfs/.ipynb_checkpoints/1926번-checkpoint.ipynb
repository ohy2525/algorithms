```python
import sys
sys.setrecursionlimit(10**6)

dx = [-1, 1, 0, 0]
dy = [0, 0, -1, 1]
def dfs(x, y):
    global size
    graph[x][y] = 0
    size += 1

    for i in range(4):
        nx = x + dx[i]
        ny = y + dy[i]
        if 0 <= nx < n and 0 <= ny < m and graph[nx][ny] == 1:
            dfs(nx, ny)

n, m = map(int, input().split())
graph = [list(map(int, input().split())) for _ in range(n)]


cnt = 0       
max_size = 0  
size = 0
for i in range(n):
    for j in range(m):
        if graph[i][j] == 1:
            size = 0
            dfs(i, j)
            max_size = max(max_size, size)      
            cnt += 1


print(cnt)
print(max_size)
```
