```python
import sys
sys.setrecursionlimit(1000000)

dx = [-1, 1, 0, 0]
dy = [0, 0, -1, 1]

def dfs(x, y):
    visited[x][y] = True
    color = graph[x][y]
    
    for i in range(4):
        nx = x + dx[i]
        ny = y + dy[i]
        
        if 0 <= nx < n and 0 <= ny < n and visited[nx][ny] == False and graph[nx][ny] == color:
            dfs(nx, ny)

n = int(input())
graph = [list(input()) for  _ in range(n)]

visited = [[False] * n for _ in range(n)]
cnt_3 = 0
for i in range(n):
    for j in range(n):
        if visited[i][j] == False:
            dfs(i,j)
            cnt_3 += 1
            
for i in range(n):
    for j in range(n):
        if graph[i][j] == "R":
            graph[i][j] = "G"     

visited = [[False] * n for _ in range(n)]            
cnt_2 = 0
for i in range(n):
    for j in range(n):
        if visited[i][j] == False:
            dfs(i,j)
            cnt_2 += 1
            
print(cnt_3, end = " ")
print(cnt_2)
```
