```python
import sys
sys.setrecursionlimit(10 ** 6)

dx = [-1, 1, 0, 0]
dy = [0, 0, -1, 1]

def dfs_W(x, y):
    if x <= -1 or x >= m or y <= -1 or y >= n:
        return False

    if graph[x][y] == "W":
        cnt_W.append(1)
        graph[x][y] = "V"
        for i in range(4):
            nx = x + dx[i]
            ny = y + dy[i]
            dfs_W(nx, ny)
        return True
        
    return False

def dfs_B(x, y):
    if x <= -1 or x >= m or y <= -1 or y >= n:
        return False

    if graph[x][y] == "B":
        cnt_B.append(1)
        graph[x][y] = "V"
        for i in range(4):
            nx = x + dx[i]
            ny = y + dy[i]
            dfs_B(nx, ny)
        return True
        
    return False


n, m = map(int,input().split())
graph = [list(map(str, input())) for _ in range(m)]

cnt_W = []
res_W = []

cnt_B = []
res_B = []

for i in range(m):
    for j in range(n):
        if dfs_W(i, j):
            res_W.append(len(cnt_W) ** 2)
            cnt_W = []

        elif dfs_B(i, j):
            res_B.append(len(cnt_B) ** 2)
            cnt_B = []

print(sum(res_W), sum(res_B))
```
