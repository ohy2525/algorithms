```python
import sys
sys.setrecursionlimit(100000)

graph = [list(map(int, input().split())) for _ in range(5)]

dx = [-1, 1, 0, 0]
dy = [0, 0, -1, 1]
num = ''
def dfs(x, y, num, cnt):
    if x < 0 or x >= 5 or y < 0 or y >= 5:
        return False
    num = num + str(graph[x][y])
    cnt += 1
    if cnt == 6:
        ans.add(num)
        return True
    for i in range(4):
        nx = x + dx[i]
        ny = y + dy[i]
        dfs(nx, ny, num, cnt)
        
ans = set()
for i in range(5):
    for j in range(5):
        dfs(i, j, num, 0)
        
print(len(ans))
```
