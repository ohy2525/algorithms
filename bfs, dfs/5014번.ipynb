https://www.acmicpc.net/problem/5014

```python
from collections import deque

f, s, g, u,d = map(int, input().split())
visited = [0 for _ in range(f + 1)]
def bfs(s):
    q = deque([s])
    visited[s] = 1
    while q:
        x = q.popleft()
        if x == g:
            print(visited[x] - 1)
            return
        for i in (x + u, x - d):
            if 0 < i <= f and not visited[i]:
                visited[i] = visited[x] + 1
                q.append(i)
    else:
        print("use the stairs")
        return

bfs(s)
```
