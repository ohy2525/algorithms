       N, M  P (1  N, M  105, 1  P  M)  ,  , TV    TV    .

  N            ai  bi (1  ai, bi  M, ai  bi) .

     .

         .      , -1 .

```python
from collections import deque
import sys
#input = sys.stdin.readline
n, m, p = map(int, input().split())

graph = [0 for i in range(m + 1)]
visited = [False for i in range(m + 1)]
for i in range(n):
    a, b = map(int, input().split())
    if graph[b] == 0:
        graph[b] = a
        
def dfs(start):
    q = deque()
    q.append(start)
    visited[start] = True
    cnt = 0
    while q:
        x = q.popleft()
        if visited[graph[x]]:
            return -1
        
        if graph[x] != 0 and not visited[graph[x]]:
            q.append(graph[x])
            visited[graph[x]] = True
            cnt += 1
        
        if graph[x] == 0:
            break
            
    return cnt

print(dfs(p))
```
