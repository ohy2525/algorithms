```python
import sys
#input = sys.stdin.readline
from collections import deque
n, m = map(int, input().split())
graph = [[] for _ in range(n + 1)]

for _ in range(m):
    a, b = map(int, input().split())
    #graph[a].append(b)
    graph[b].append(a)
    
def bfs(start):
    cnt = 1
    queue = deque()
    queue.append(start)
    visited = [False] * (n + 1)
    visited[start] = True
    
    while queue:
        x = queue.popleft()
        for i in graph[x]:
            if not visited[i]:
                visited[i] = True
                cnt += 1
                queue.append(i)
                
    return cnt

maxcount = 1
ans = []
for i in range(1, n + 1):
    cnt = bfs(i)
    if cnt > maxcount:
        maxcount = cnt
        ans.clear()
        ans.append(i)
    elif cnt == maxcount:
        ans.append(i)
        
print(*ans)
```
