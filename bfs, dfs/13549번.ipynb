```python
from collections import deque
n, k = map(int, input().split())
visited = [-1 for _ in range(100001)]

def bfs(x):
    q = deque()
    q.append(x)
    visited[x] = 0
    while q:
        x = q.popleft()
        if x == k:
            print(visited[x])
            return
        if 0 <= x - 1 <= 100000 and visited[x - 1] == -1:
            visited[x - 1] = visited[x] + 1
            q.append(x - 1)
        if 0 <= 2*x <= 100000 and visited[2*x] == -1:
            visited[2*x] = visited[x]
            q.append(2*x)
        if 0 <= x + 1 <= 100000 and visited[x + 1] == -1:
            visited[x + 1] = visited[x]
            q.append(x + 1)
            
bfs(n)
```

```python
from collections import deque

n, k = map(int, input().split())  # n:   , k:   
q = deque()
q.append(n) 
visited = [-1 for _ in range(100001)]
visited[n] = 0

while q:
    s = q.popleft()
    if s == k:
        print(visited[s])
        break
    if 0 <= s-1 < 100001 and visited[s-1] == -1:
        visited[s-1] = visited[s] + 1
        q.append(s-1)
    if 0 < s*2 < 100001 and visited[s*2] == -1:
        visited[s*2] = visited[s]
        q.appendleft(s*2)  # 2*s        
    if 0 <= s+1 < 100001 and visited[s+1] == -1:
        visited[s+1] = visited[s] + 1
        q.append(s+1)
```
