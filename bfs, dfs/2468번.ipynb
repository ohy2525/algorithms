```python
import sys
import copy
sys.setrecursionlimit(10**6)
input = sys.stdin.readline

n = int(input())
graph = []
max_num = 0
for i in range(n):
    graph.append(list(map(int, input().split())))
    for j in range(n):
        if graph[i][j] > max_num:
            max_num = graph[i][j]

dx = [-1, 1, 0, 0]
dy = [0, 0, 1, -1]

def dfs(x, y):
    new_graph[x][y] = 0
    
    for i in range(4):
        nx = x + dx[i]
        ny = y + dy[i]
        
        if 0 <= nx < n and 0 <= ny < n and new_graph[nx][ny] == 1:
            dfs(nx, ny)
            
ans = 0     

for i in range(max_num):
    
    new_graph = copy.deepcopy(graph)
    
    for j in range(n):
        for k in range(n):
            if new_graph[j][k] > i:
                new_graph[j][k] = 1
            else :
                new_graph[j][k] = 0
                
    count = 0
    for x in range(n):
        for y in range(n):
            if new_graph[x][y] == 1:
                dfs(x, y)
                count += 1
                
    ans = max(ans, count)
        
print(ans)
```
