https://www.acmicpc.net/problem/2583

```python
import sys
sys.setrecursionlimit(1000000)

m, n, k = map(int, input().split())
graph = [[0] * n for _ in range(m)]
for _ in range(k):
    x1, y1, x2, y2 = map(int, input().split())
    for i in range(y1, y2):
        for j in range(x1, x2):
            graph[i][j] = 1
            
dx = [-1, 1, 0, 0]
dy = [0, 0, -1, 1]

def dfs(y, x, cnt):
    graph[y][x] = 1
    for i in range(4):
        nx = x + dx[i]
        ny = y + dy[i]
        if 0 <= nx < n and 0 <= ny < m and graph[ny][nx] == 0:
            cnt = dfs(ny, nx, cnt + 1)
    return cnt

ans = []
for i in range(m):
    for j in range(n):
        if graph[i][j] == 0 :
            ans.append(dfs(i, j, 1))
            
print(len(ans))
print(*sorted(ans))
```
