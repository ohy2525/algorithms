```python
a, b = map(int, input().split())
cnt = 1
while b != a:
    cnt += 1
    temp = b
    if b % 10 == 1:
        b = b // 10
    elif b % 2 == 0:
        b = b // 2
    
    if temp == b:
        print(-1)
        break
else :
    print(cnt)
```

```python
from collections import deque
a, b = map(int, input().split())
queue = deque()
queue.append((a, 1))

while queue:
    num, cnt = queue.popleft()
    if num > b:
        continue
    if num == b:
        print(cnt)
        break
    
    queue.append((2*num, cnt + 1))
    queue.append((int(str(num) + '1'), cnt + 1))
else :
    print(-1)
```
