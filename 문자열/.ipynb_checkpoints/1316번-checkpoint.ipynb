```python
import itertools
n = int(input())
res = 0
for _ in range(n):
    s= input()
    v = list(s)
    g = itertools.groupby(v)
    if len(set(v)) == len([k for k, v in g]) :
        res +=1
print(res)
```

```python
res = 0
for i in range(int(input())):
    word = input()
    if list(word) == sorted(word, key=word.find):
        res += 1
print(res)
```
