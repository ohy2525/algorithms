https://www.acmicpc.net/problem/1181

```python
n = int(input())
array = []
for _ in range(n):
    y = input()
    x = len(y)
    array.append([x, y])

new_array = []
for word in array:
    if word not in new_array:
        new_array.append(word)
new_array.sort()

for x, y in new_array:
    print(y)
```
