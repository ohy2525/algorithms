https://www.acmicpc.net/problem/1940

```python
import sys
#input = sys.stdin.readline
n = int(input())
m = int(input())
array = list(map(int, input().split()))

array.sort()

start = 0
end = n -1
ans = 0

while start < end:
    tmp = array[start] + array[end]
    if tmp == m:
        ans += 1
        start += 1
    elif tmp > m:
        end -= 1
    elif tmp < m:
        start += 1
        
print(ans)
```
