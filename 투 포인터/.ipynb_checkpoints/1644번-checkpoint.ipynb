https://www.acmicpc.net/problem/1644

```python
n = int(input())
array = []
prime_number = [False, False] + [True] * (n - 1)
for i in range(2, int(n**0.5) + 1):
    if prime_number[i]:
        prime_number[i*2::i] = [False] * ((n - i) // i)
for i in range(n + 1):
    if prime_number[i]:
        array.append(i)
        
cnt, sum_ = 0, 0
start, end = 0, 0
while True:
    if sum_ == n:
        cnt += 1
    
    if sum_ > n:
        sum_ -= array[start]
        start += 1
    elif end == len(array):
        break
    else:
        sum_ += array[end]
        end += 1
    print(sum_)
    print(start)
    print(end)
    print("----------------")
        
print(cnt)
```
