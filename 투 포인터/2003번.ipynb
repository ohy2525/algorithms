https://www.acmicpc.net/problem/2003

```python
n, m = map(int, input().split())
array = list(map(int, input().split()))
ans = 0
start, end = 0, 1
while end <= n and start <= end:
    total = sum(array[start:end])
    if total == m:
        ans += 1
        end += 1
    elif total < m:
        end += 1
    else:
        start += 1
print(ans)
```
