```python
import sys
n, k = map(int, input().split())
grade = list(map(int, input().split()))

grade_sum = [0]
for i in range(n):
    grade_sum.append(grade_sum[i] + grade[i])

for _ in range(k):
    s, e = map(int, input().split())
    print(f"{(grade_sum[e] - grade_sum[s - 1]) / (e - s + 1):.2f}")
```
