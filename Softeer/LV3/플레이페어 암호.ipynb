```python
def find_index(c):
    global playfair
    for i in range(5):
        for j in range(5):
            if playfair[i][j] == c:
                return i, j
    
msg = input()
key = list(input())

alphabet = [chr(i + 65) for i in range(26)]
alphabet.remove('J')

playfair = []
for k in key:
    if k in alphabet:
        playfair.append(k)
        alphabet.remove(k)
playfair.extend(alphabet)
playfair = [playfair[i:i + 5] for i in range(0, 25, 5)]

before = []
i = 0
while i < len(msg):
    if i + 1 < len(msg):
        if msg[i] != msg[i + 1]:
            before.append(msg[i:i+2])
            i += 2
        elif msg[i] == msg[i+1]:
            if msg[i] == 'X':
                before.append(msg[i] + 'Q')
                i += 1
            else:
                before.append(msg[i] + 'X')
                i += 1
    else:
        before.append(msg[i] + 'X')
        i += 1

after = ""
for w in before:
    x1, y1 = find_index(w[0])
    x2, y2 = find_index(w[1])
    if x1 == x2:
        after += playfair[x1][(y1 + 1) % 5]
        after += playfair[x2][(y2 + 1) % 5]
    elif y1 == y2:
        after += playfair[(x1 + 1) % 5][y1]
        after += playfair[(x2 + 1) % 5][y2]
    else:
        after += playfair[x1][y2]
        after += playfair[x2][y1]

print(after)
```
