```python
import sys
from itertools import permutations

n, m, k = map(int, input().split())
arr = list(map(int, input().split()))

cases = permutations(arr)
answer = sys.maxsize

for case in cases:
    case = list(case)
    cnt = 0
    weight = 0
    idx = 0
    tmp_answer = 0
    
    while cnt != k:
        if weight + case[idx] <= m:
            weight += case[idx]
            case.append(case[idx])
            idx += 1
        else:
            tmp_answer += weight
            cnt += 1
            weight = 0
    answer = min(answer, tmp_answer)

print(answer)
```
