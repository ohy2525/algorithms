```python
import sys
input = sys.stdin.readline
W, N = map(int, input().split())
arr = []
for _ in range(N):
    m, p = map(int, input().split())
    arr.append((m, p))
arr.sort(reverse = True, key = lambda x : x[1])

answer = 0
for i in range(N):
    if W > arr[i][0]:
        answer += arr[i][0] * arr[i][1]
        W -= arr[i][0]

    else:
        answer += W * arr[i][1]
        break

print(answer)
```
