```python
import sys
input = sys.stdin.readline

def dfs(x, y):
    global cnt
    visited[x][y] = True
    cnt += 1
    for i in range(4):
        nx = x + dx[i]
        ny = y + dy[i]
        if 0 <= nx < n and 0 <= ny < n and not visited[nx][ny] and arr[nx][ny] == '1':
            dfs(nx, ny)
            


n = int(input())
arr = []
for _ in range(n):
    arr.append(list(input()))

dx = [-1, 1, 0, 0]
dy = [0, 0, -1, 1]
answer = []
visited = [[False] * n for _ in range(n)]
for i in range(n):
    for j in range(n):
        if arr[i][j] == '1' and not visited[i][j]:
            cnt = 0
            dfs(i, j)
            answer.append(cnt)

answer.sort()
print(len(answer))
for c in answer:
    print(c)
```
