https://www.acmicpc.net/problem/1874

```python
n = int(input())
ans = []
check = []
flag = True
count = 1
for i in range(n):
    num = int(input())
    while count <= num :
        check.append(count)
        ans.append('+')
        count += 1
        
    if check[-1] == num :
        check.pop()
        ans.append('-')
    else:
        flag = False
        
if flag == False:
    print('NO')
else :
    for i in ans:
        print(i)
```
