```python
import sys
#input = sys.stdin.readline
n = int(input())
dic = {}
for _ in range(n):
    extension = input().split('.')[1]
    try:
        dic[extension] += 1
    except:
        dic[extension] = 1

sort = sorted(dic.items())

for k, v in sort:
    print(k.rstrip(), v)
```
