https://www.acmicpc.net/problem/4948

```python
e = 123456*2
a = [False,False] + [True]*(e-1)
primes=[]

for i in range(2,e+1):
  if a[i]:
    primes.append(i)
    for j in range(2*i, e+1, i):
        a[j] = False

        
while True:
    i = int(input())
    if i == 0 :
        break
    print(sum([i<x and x<=2*i for x in primes]))
```
