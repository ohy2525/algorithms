```python
from collections import Counter
n = int(input())
num_list = []
for _ in range(n):
    num_list.append(int(input()))
num_list.sort()
#평균
print(round(sum(num_list)/n))

#중앙값
print(num_list[n//2])

#최빈값
cnt = Counter(num_list).most_common(2)
if len(cnt)==2:
    if cnt[0][1] == cnt[1][1]:
        print(cnt[1][0])
    else :
        print(cnt[0][0])
        
else :
    print(cnt[0][0])
    
#범위
print(num_list[-1] - num_list[0])
```
