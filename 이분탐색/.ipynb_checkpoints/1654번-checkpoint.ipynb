https://www.acmicpc.net/problem/1654

```python
import sys
#input = sys.stdin.readline
k, n = map(int, input().split())
array = []
for _ in range(k):
    array.append(int(input()))
    
array.sort()

start = 1
end = max(array)

while (start <= end):
    mid = (start + end) // 2
    cnt = 0
    for i in array:
        cnt += i // mid
    if cnt >= n:
        start = mid + 1
    else:
        end = mid -1
        
print(end)
```
