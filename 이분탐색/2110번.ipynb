```python
import sys
#input = sys.stdin.readline

n, c = map(int, input().split())
array = []
for _ in range(n):
    array.append(int(input()))
array.sort()
start = 1
end = array[-1] - array[0]
while (start <= end) :
    mid = (start + end) // 2
    current = array[0]
    cnt = 1
    for i in range(1, n):
        if current + mid <= array[i]:
            current = array[i]
            cnt += 1
    if cnt >= c:
        start = mid + 1
    else:
        end = mid -1
print(end)
```
