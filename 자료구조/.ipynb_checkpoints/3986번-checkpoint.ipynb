https://www.acmicpc.net/problem/3986

```python
n = int(input())
ans = 0

for _ in range(n):
    tmp = input().rstrip()
    stack = []
    
    for i in range(len(tmp)):
        if stack and tmp[i] == stack[-1]:
            stack.pop()
            
        else:
            stack.append(tmp[i])
            
    if not stack:
        ans += 1
print(ans)
```
