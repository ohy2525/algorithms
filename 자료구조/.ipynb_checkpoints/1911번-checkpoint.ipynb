https://www.acmicpc.net/problem/1911

```python
from collections import deque
n, l = map(int, input().split())
holes = []
for _ in range(n):
    holes.append(list(map(int, input().split())))

holes.sort()

answer = 0
for i in range(n):
    start, end = holes[i][0], holes[i][1]
    length = end - start
    answer += length // l
    
    if length % l != 0:
        if i != n - 1:
            if end + (l - length % l) >= holes[i + 1][0]:
                holes[i + 1][0] = end + (l - length % l)
            answer += 1
            
        else:
            answer += 1

print(answer)     
```
