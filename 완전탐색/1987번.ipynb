https://www.acmicpc.net/problem/1987

```python
r, c = map(int, input().split())
a = [input() for _ in range(r)]

ans = 0
visited = [False for _ in range(26)]

dx = [-1, 1, 0, 0]
dy = [0, 0, -1, 1]

def backtracking(x, y, cnt):
    global ans
    ans = max(ans, cnt)
    visited[ord(a[x][y]) - ord("A")] = True
    
    for i in range(4):
        nx = x + dx[i]
        ny = y + dy[i]
        
        if 0 <= nx < r and 0 <= ny < c and not visited[ord(a[nx][ny]) - ord("A")]:
            backtracking(nx, ny, cnt + 1)
            
    visited[ord(a[x][y]) - ord("A")] = False

backtracking(0, 0, 1)

print(ans)
```
