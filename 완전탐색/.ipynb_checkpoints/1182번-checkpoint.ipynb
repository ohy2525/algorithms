```python
import sys
from itertools import combinations
#input = sys.stdin.readline
n, s = map(int, input().split())
array = list(map(int, input().split()))
cnt = 0
for i in range(1, n + 1):
    comb = combinations(array, i)
    
    for x in comb:
        if sum(x) == s:
            cnt += 1
            
print(cnt)
```

```python
n, s = map(int, input().split())
array = list(map(int, input().split()))
cnt = 0

def sol(idx, sub_sum):
    global cnt
    
    if idx >= n:
        return
    
    sub_sum += array[idx]
    
    if sub_sum == s:
        cnt += 1
        
    #  
    sol(idx + 1, sub_sum)
    #   
    sol(idx + 1, sub_sum - array[idx])
    
sol(0, 0)
print(cnt)
```
