https://www.acmicpc.net/problem/1449

```python
n, l = map(int, input().split())
array = list(map(int, input().split()))
array.sort()

start = array[0]
end = array[0] + l
cnt = 1

for i in range(n):
    if start <= array[i] < end:
        continue
    else:
        start = array[i]
        end = array[i] + l
        cnt += 1
print(cnt)
```
