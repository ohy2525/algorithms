https://www.acmicpc.net/problem/1978

```python
n = int(input())
numbers = map(int, input().split())
res = 0
for num in numbers:
    check = 0
    if num > 1 :
        for i in range(2, num):  
            if num % i == 0:
                check += 1  
        if check == 0:
            res += 1  
print(res)
```
