https://www.acmicpc.net/problem/13458

```python
n = int(input())
num = list(map(int, input().split()))
b, c = map(int, input().split())
ans = 0
for i in num:
    i -= b
    ans += 1
    if i > 0:
        if i % c != 0:
            ans += (i // c + 1)
        else:
            ans += i // c
        
print(ans)
```
