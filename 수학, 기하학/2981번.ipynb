```python
from math import gcd
from math import sqrt
```

```python
n = int(input())
num_list = []
interval = []
ans = []
for _ in range(n):
    num_list.append(int(input()))
    
num_list.sort()

for i in range(1, n):
    interval.append(num_list[i] - num_list[i-1])

prev = interval[0]
for i in range(1, len(interval)):
    prev = gcd(prev, interval[i])
    
for i in range(2, int(sqrt(prev)) + 1):
    if prev % i == 0:
        ans.append(i)
        ans.append(prev//i)
ans.append(prev)
ans.sort()
print(*ans)
```
