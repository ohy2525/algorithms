```python
import sys
import heapq
INF = int(1e9)

start_x, start_y = map(float, sys.stdin.readline().split())
end_x, end_y = map(float, sys.stdin.readline().split())
n = int(sys.stdin.readline())

def length(x, y, new_x, new_y):
    return((x-new_x)**2 + (y-new_y)**2)**0.5

def dijkstra(start, graph):
    distance = [INF for _ in range(len(canon))]
    q = []
    
    distance[start] = 0
    heapq.heappush(q, (distance[start], start))
    
    while q:
        dist, now = heapq.heappop(q)
        if distance[now] < dist:
            continue
        for p, c in enumerate(graph[now]):
            cost = dist + c
            if cost < distance[p]:
                distance[p] = cost
                heapq.heappush(q, (cost, p))
    return distance

canon = []
canon.append((start_x, start_y))

while n :
    n -= 1
    canon_x, canon_y = map(float, sys.stdin.readline().split())
    canon.append((canon_x, canon_y))
canon.append((end_x, end_y))

time = [[] for _ in range (len(canon))]

for i in range(len(canon)):
    for j in range(len(canon)):
        if i == 0 or i ==len(canon)-1:
            time[i].append(length(canon[i][0], canon[i][1], canon[j][0], canon[j][1])/5)
        else :
            time[i].append(abs(length(canon[i][0], canon[i][1], canon[j][0], canon[j][1])-50)/5+2)
     

print(dijkstra(0, time)[len(canon) -1])
```
