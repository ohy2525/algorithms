https://www.acmicpc.net/problem/9020

```python
primes = [0 for i in range(10001)]
primes[1] = 1
for i in range(2, 98):
    for j in range(i * 2, 10001, i):
        primes[j] = 1
case = int(input())
for i in range(case):
    n = int(input())
    b = n // 2
    for j in range(b, 1, -1):
        if primes[n - j] == 0 and primes[j] == 0:
            print(j, n - j)
            break
            
```
