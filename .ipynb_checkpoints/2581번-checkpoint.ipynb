https://www.acmicpc.net/problem/2581

```python
m = int(input())
n = int(input())

res = []
check = 0
if m == 1:
    m = 2
for i in range(m, n+1):
    for j in range(2, i):
        if i % j == 0:
            check +=1
            if check != 0:
                break
    if check == 0:
        res.append(i)
    check =0
if len(res) == 0:
    print(-1)
else :
    print(sum(res))
    print(res[0])
```
