https://www.acmicpc.net/problem/9012

```python
n = int(input())

for i in range(n):
    VPS = input()
    sum = 0
    for a in VPS :
        if a == '(' :
            sum += 1
        elif a == ')':
            sum -= 1
        
        if sum < 0 :
            print('NO')
            break
    if sum == 0 :
        print('YES')

    elif sum > 0 :
        print('NO')
```
